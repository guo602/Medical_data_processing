```python
import matplotlib.pyplot as plt

# 假设你有一个名为 data 的列表，其中包含了你的数据
data = [[1, 2, 3],
        [1, 2, 3],
        [1, 2, 3],
  ]  # 这里只是一个示例数据
PLOT_PER_ROW = 5
SINGLW_PERIOD_PLOT_SIZE = (18,6)

# 计算子图的行数和列数
num_plots = len(data)
num_rows = int(num_plots/PLOT_PER_ROW) + int(num_plots%PLOT_PER_ROW > 0)  # 每行最多显示三个子图
num_cols = min(PLOT_PER_ROW, num_plots)  # 每列最多显示三个子图


# 创建画布和子图
fig, axes = plt.subplots(num_rows, num_cols, figsize=SINGLW_PERIOD_PLOT_SIZE)

# 遍历数据并绘制子图
for i, ax in enumerate(axes.flat):
    if i < num_plots:
        ax.plot([1,2,3],data[i])
    else:
        # 如果子图数量超过实际数据数量，则隐藏多余的子图
        ax.axis('off')

# 自动调整子图之间的间距
plt.tight_layout()

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt

# 创建一个大小为(8, 6)的图形，并以2行1列的网格布局
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# 在第一个子图中绘制数据
axs[0].set_title('Subplot 1')
axs[0].plot([1, 2, 3], [4, 5, 6])

# 在第二个子图中绘制数据
axs[1].set_title('Subplot 2')
axs[1].plot([1, 2, 3], [6, 5, 4])

# 调整子图之间的间距
plt.tight_layout()

# 显示图形
plt.show()
```
